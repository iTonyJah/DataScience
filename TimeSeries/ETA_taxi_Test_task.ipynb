```python
import pandas as pd

df = pd.read_csv('data/taxi.csv')
df.head()
```

```python
df['datetime'].min(), df['datetime'].max(), df.shape
```

```python
df.dtypes
```

```python
df['datetime'] = pd.to_datetime(df['datetime'], format='%d/%m/%Y %H:%M')
```

```python
df.dtypes
```

```python
import matplotlib.pyplot as plt

plt.plot(df.set_index('datetime'))
```

```python
import plotly.express as px

px.line(df.set_index('datetime'))
```

```python
import statsmodels.api as sm

decompose = sm.tsa.seasonal_decompose(df.set_index('datetime'), model='additive', period=6 * 24)

fig = decompose.plot()

fig.set_size_inches(15, 8)

fig.show()
```

```python
import statsmodels.api as sm

decompose = sm.tsa.seasonal_decompose(df.set_index('datetime'), model='additive', period=6 * 24 * 30)

fig = decompose.plot()

fig.set_size_inches(15, 8)

fig.show()
```
